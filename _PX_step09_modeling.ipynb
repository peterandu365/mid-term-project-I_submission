## Logistic Regression


LogisticRegression does not accept missing values encoded as NaN natively. For supervised learning, we might want to consider sklearn.ensemble.HistGradientBoostingClassifier and Regressor which accept missing values encoded as NaNs natively. Alternatively, it is possible to preprocess the data, for instance by using an imputer transformer in a pipeline or drop samples with missing values. See https://scikit-learn.org/stable/modules/impute.html we can find a list of all estimators that handle NaN values at the following page: https://scikit-learn.org/stable/modules/impute.html#estimators-that-handle-nan-values

```python
columns_X_plus = columns_X

df_m = df.copy()
print('size before dropna:', df_m.shape)

# Drop the rows with NaN values only in the specified columns from columns_X_plus
df_m.dropna(subset=columns_X_plus, inplace=True)

print('size after dropna:', df_m.shape)


# df_m = df.copy()
# df_m.drop(columns_useless, axis=1, inplace=True)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

#  the dataframe is loaded as df_m

# Create the binary target variable
df_m['delayed'] = df_m[['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']].sum(axis=1)
df_m['delayed'] = np.where(df_m['delayed'] > 0, 1, 0)

# Define the independent and dependent variables
_columns_X = columns_X
_columns_y = 'delayed'

X = df_m[_columns_X]
y = df_m[_columns_y]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create and fit the logistic regression model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg.predict(X_test)

# Calculate the accuracy, confusion matrix, and classification report
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)

# ======================  Print the results  ======================
# Save reports and conclusion into a txt file
with open(MODEL_RESULT_PATH + "01_delay_reports_and_conclusion_LogisticRegression.txt", "w") as f:
    f.write(f"Accuracy: {accuracy}\n")
    f.write("Confusion Matrix:\n")
    f.write(np.array2string(conf_matrix, separator=', '))
    f.write("\nClassification Report:\n")
    f.write(class_report)

# Save the predicted classifications in a csv file
test_results = X_test.copy()
test_results['true_delayed'] = y_test
test_results['predicted_delayed'] = y_pred
test_results.to_csv(MODEL_RESULT_PATH + "01_delay_predicted_classifications_LogisticRegression.csv", index=False)

```

PCA (Principal Component Analysis) and feature scaling can be used in conjunction with most machine learning techniques. However, some algorithms are less sensitive to feature scaling, and some inherently handle it internally. Here's a summary of how PCA and feature scaling relate to the mentioned techniques:

* Linear/Logistic/Multinomial Logistic Regression: Both PCA and feature scaling can be applied. Feature scaling is particularly important for these methods, as they are sensitive to the scale of input features.

* Naive Bayes: PCA can be applied. However, feature scaling is not necessary, as the algorithm works with probabilities and is not sensitive to the scale of input features.

* Random Forest: PCA can be applied, but it may not always lead to better performance since trees-based methods are generally robust to feature scaling. Feature scaling is not necessary, as decision trees are not sensitive to the scale of input features.

* SVM (Support Vector Machines): Both PCA and feature scaling can be applied. Feature scaling is particularly important for SVMs, as they are sensitive to the scale of input features.

* XGBoost: PCA can be applied, but it may not always lead to better performance since trees-based methods are generally robust to feature scaling. Feature scaling is not necessary, as decision trees are not sensitive to the scale of input features.

* Ensemble methods: The applicability of PCA and feature scaling depends on the base estimator used in the ensemble. For example, if we use an ensemble of logistic regression models, we should apply feature scaling. If we use an ensemble of decision trees or tree-based methods, feature scaling may not be necessary.

In summary, PCA can be applied to all the mentioned techniques, while feature scaling is particularly important for linear models (linear/logistic/multinomial logistic regression) and SVM. For other methods, especially tree-based models, feature scaling is not as critical.

### Use PCA to reduce the number of features:

The n_components parameter in PCA is set to a float value (0.95) instead of an integer. This means that PCA will choose the number of components such that the amount of explained variance is greater than or equal to the specified percentage (95% in this case).

By setting n_components=0.95, PCA will automatically select the minimum number of components required to retain at least 95% of the total variance in the dataset.

If explicitly set the number of components, we can use an integer value for n_components. For example, setting n_components=5 would keep the first five principal components. However, using a float value allows PCA to adaptively choose the number of components based on the data's explained variance.

```python
# drop na should be done after the train/test split, otherwise the test set will have less rows.
columns_X_plus = columns_X + ['cancelled']
df_m = df.copy()
print('size before dropna:', df_m.shape)

# Drop the rows with NaN values only in the specified columns from columns_X_plus
df_m.dropna(subset=columns_X_plus, inplace=True)

print('size after dropna:', df_m.shape)


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

#  the dataframe is loaded as df_m
_columns_X = columns_X
_columns_y = 'cancelled'

# print('size before dropna:', df_m[columns_X].shape)
X = df_m[columns_X].dropna()
y = df_m[_columns_y]
# print('size after dropna:', X.shape)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Apply feature scaling using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply PCA to reduce the number of features
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Define the parameter grid for GridSearchCV
param_grid = {
    'learning_rate': [0.01, 0.1, 1],
    'max_iter': [100, 500, 1000],
    'max_depth': [3, 5, 7],
    'min_samples_leaf': [5, 10, 20],
    'l2_regularization': [0, 0.1, 1]
}

# Initialize HistGradientBoostingClassifier and GridSearchCV
hgbc = HistGradientBoostingClassifier()
grid_search = GridSearchCV(estimator=hgbc, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit the GridSearchCV model
grid_search.fit(X_train_pca, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Make predictions on the test set
y_pred = grid_search.predict(X_test_pca)

# Calculate the accuracy, confusion matrix, and classification report
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)

# Save reports and conclusion into a txt file
with open(MODEL_RESULT_PATH + "04c_cancel_reports_and_conclusion_HistGradientBoostingClassifier_PCA_GridSearch.txt", "w") as f:
    f.write(f"Accuracy: {accuracy}\n")
    f.write("Confusion Matrix:\n")
    f.write(np.array2string(conf_matrix, separator=', '))
    f.write("\nClassification Report:\n")
    f.write(class_report)

# Save the predicted classifications in a csv file
test_results = X_test.copy()
test_results['true_cancelled'] = y_test
test_results['predicted_cancelled'] = y_pred
test_results.to_csv(MODEL_RESULT_PATH + "04c_cancel_predicted_classifications_HistGradientBoostingClassifier_PCA_GridSearch.csv", index=False)
```

## MultiOutputRegressor
